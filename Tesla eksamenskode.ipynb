```python
arrNewsHeadandAbstract = []
for i in range(0,100,1):
    for p in information[i]["response"]["docs"]:
        print(p["headline"]["main"])
        arrNewsHeadandAbstract.append(p["headline"]["main"])
        print(p["abstract"])
        arrNewsHeadandAbstract.append(p["abstract"])
        date_of_article = (p["pub_date"])
        print(date_of_article[0:10])
        arrNewsHeadandAbstract.append(date_of_article[0:10])
        print("break")
        
print()
```

```python
dfNewsText = pd.DataFrame(arrNewsHeadandAbstract, columns=['NewsText'])
```

```python
df1=dfNewsText.iloc[0:len(dfNewsText):3] #headlinemain
df2=dfNewsText.iloc[1:len(dfNewsText):3] #headlineabstract
df3=dfNewsText.iloc[2:len(dfNewsText):3] #headlinedate
df1.reset_index(drop=True, inplace=True)
df2.reset_index(drop=True, inplace=True)
df3.reset_index(drop=True, inplace=True)
df1.rename(columns = {'NewsText':'Headline'})
df2.rename(columns = {'NewsText':'Abstract'})
df3.rename(columns = {'NewsText':'Date'})
```

```python
Finaldf = pd.concat([df1, df2, df3], axis=1)

# Reset the index if needed
Finaldf.reset_index(drop=True, inplace=True)

# If you want to rename the columns to make it clear which dataframe each column came from
Finaldf.columns = ['Headline', 'Abstract', 'Date']

Finaldf
```
